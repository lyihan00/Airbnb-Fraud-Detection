```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, precision_recall_curve, roc_auc_score
import warnings
import os

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
# Import Datasets for Train and Test
train_vanilla = pd.read_csv('train_vanilla.csv')
train_undersampled = pd.read_csv('train_undersampled.csv')
train_synthetic = pd.read_csv('train_synthetic.csv')

test_df = pd.read_csv('test_data.csv')
```

# Split Datasets into features and target
Target column: `fraud`

## Split Vanilla Train Dataset

```python
train_vanilla.head()
```

```python
X_train_vanilla = train_vanilla.copy().drop(columns='fraud')
Y_train_vanilla = train_vanilla['fraud']
```

## Split Undersampled Train Dataset

```python
train_undersampled.head()
```

```python
X_train_undersampled = train_undersampled.copy().drop(columns='fraud')
Y_train_undersampled = train_undersampled['fraud']
```

## Split Synthetic Train Dataset

```python
train_synthetic.head()
```

```python
X_train_synthetic = train_synthetic.copy().drop(columns='fraud')
Y_train_synthetic = train_synthetic['fraud']
```

## Splitting Test Set

```python
test_df.head()
```

```python
X_test = test_df.copy().drop(columns='fraud')
Y_test = test_df['fraud']
```

# Decision Tree Classifer
Decision Tree is a supervised learning algorithm, when classification tasks can be utilized without assuming any relationship between independent and dependent variables, capturing complex non-linear relationships between predictors and targets.

* Works well with imbalanced dataset

## Fitting Decision Tree Classifer on Train Sets
### Train Sets: Vanilla, Undersampled, Synthetic

```python
# Fit Decision Tree on Vanilla
dt_vanilla = tree.DecisionTreeClassifier(random_state=42)
dt_vanilla.fit(X_train_vanilla.values, Y_train_vanilla)
```

```python
# Fit Decision Tree on Undersampled
dt_undersampled = tree.DecisionTreeClassifier(random_state=42)
dt_undersampled.fit(X_train_undersampled.values, Y_train_undersampled)
```

```python
# Fit Decision Tree on Synthetic
dt_synthetic = tree.DecisionTreeClassifier(random_state=42)
dt_synthetic.fit(X_train_synthetic.values, Y_train_synthetic)
```

## Cross-Validate using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [4, 8, 12, None],
    'min_samples_split': [1, 2, 5],
    'min_samples_leaf': [1, 2, 4]
}
```

### Vanilla Train Set Decision Tree

```python
vanilla_grid_search = GridSearchCV(dt_vanilla, param_grid, cv=5, scoring='accuracy', verbose=1)

vanilla_grid_search.fit(X_train_vanilla, Y_train_vanilla)

vanilla_best_params = vanilla_grid_search.best_params_
print(f"Best Hyperparameters: {vanilla_best_params}")

vanilla_best_score = vanilla_grid_search.best_score_
print(f"Best Accuracy Score: {vanilla_best_score}")

dt_tuned_vanilla = vanilla_grid_search.best_estimator_
```

### Undersampled Train Set Decision Tree

```python
undersampled_grid_search = GridSearchCV(dt_undersampled, param_grid, cv=5, scoring='accuracy', verbose=1)

undersampled_grid_search.fit(X_train_undersampled, Y_train_undersampled)

undersampled_best_params = undersampled_grid_search.best_params_
print(f"Best Hyperparameters: {undersampled_best_params}")

undersampled_best_score = undersampled_grid_search.best_score_
print(f"Best Accuracy Score: {undersampled_best_score}")

dt_tuned_undersampled = undersampled_grid_search.best_estimator_
```

### Synthetic Train Set Decision Tree

```python
synthetic_grid_search = GridSearchCV(dt_synthetic, param_grid, cv=5, scoring='accuracy', verbose=1)

synthetic_grid_search.fit(X_train_synthetic, Y_train_synthetic)

synthetic_best_params = synthetic_grid_search.best_params_
print(f"Best Hyperparameters: {synthetic_best_params}")

synthetic_best_score = synthetic_grid_search.best_score_
print(f"Best Accuracy Score: {synthetic_best_score}")

dt_tuned_synthetic = synthetic_grid_search.best_estimator_
```

## Predict on Test set
Predict on test set based on Decision Tree classifier with tuned hyperparameters determined by GridSearchCV.

### Vanilla Trained

```python
pred_vanilla = dt_tuned_vanilla.predict(X_test)

vanilla_cf = confusion_matrix(Y_test, pred_vanilla)
print(f"Confusion Matrix for Decision Tree trained on Vanilla: \n{vanilla_cf}")

vanilla_acc = accuracy_score(Y_test, pred_vanilla)
print(f"Accuracy Score for Decision Tree trained on Vanilla = {vanilla_acc*100:.2f}%")

vanilla_pre = precision_score(Y_test, pred_vanilla)
print(f"Precision Score for Decision Tree trained on Vanilla = {vanilla_pre*100:.2f}%")

vanilla_rec = recall_score(Y_test, pred_vanilla)
print(f"Recall Score for Decision Tree trained on Vanilla = {vanilla_rec*100:.2f}%")

vanilla_f1 = f1_score(Y_test, pred_vanilla)
print(f"F1 Score for Decision Tree trained on Vanilla = {vanilla_f1*100:.2f}%")

vanilla_roc_auc = roc_auc_score(Y_test, pred_vanilla)
print(f"ROC AUC Score for Decision Tree trained on Vanilla = {vanilla_roc_auc*100:.2f}%")
```

### Undersampled Trained

```python
pred_undersampled = dt_tuned_undersampled.predict(X_test)

undersampled_cf = confusion_matrix(Y_test, pred_undersampled)
print(f"Confusion Matrix for Decision Tree trained on Undersampled: \n{undersampled_cf}")

undersampled_acc = accuracy_score(Y_test, pred_undersampled)
print(f"Accuracy Score for Decision Tree trained on Undersampled = {undersampled_acc*100:.2f}%")

undersampled_pre = precision_score(Y_test, pred_undersampled)
print(f"Precision Score for Decision Tree trained on Undersampled = {undersampled_pre*100:.2f}%")

undersampled_rec = recall_score(Y_test, pred_undersampled)
print(f"Recall Score for Decision Tree trained on Undersampled = {undersampled_rec*100:.2f}%")

undersampled_f1 = f1_score(Y_test, pred_undersampled)
print(f"F1 Score for Decision Tree trained on Undersampled = {undersampled_f1*100:.2f}%")

undersampled_roc_auc = roc_auc_score(Y_test, pred_undersampled)
print(f"ROC AUC Score for Decision Tree trained on Undersampled = {undersampled_roc_auc*100:.2f}%")
```

### Synthetic Trained

```python
pred_synthetic = dt_tuned_synthetic.predict(X_test)

synthetic_cf = confusion_matrix(Y_test, pred_synthetic)
print(f"Confusion Matrix for Decision Tree trained on Synthetic: \n{synthetic_cf}")

synthetic_acc = accuracy_score(Y_test, pred_synthetic)
print(f"Accuracy Score for Decision Tree trained on Synthetic = {synthetic_acc*100:.2f}%")

synthetic_pre = precision_score(Y_test, pred_synthetic)
print(f"Precision Score for Decision Tree trained on Synthetic = {synthetic_pre*100:.2f}%")

synthetic_rec = recall_score(Y_test, pred_synthetic)
print(f"Recall Score for Decision Tree trained on Synthetic = {synthetic_rec*100:.2f}%")

synthetic_f1 = f1_score(Y_test, pred_synthetic)
print(f"F1 Score for Decision Tree trained on Synthetic = {synthetic_f1*100:.2f}%")

synthetic_roc_auc = roc_auc_score(Y_test, pred_synthetic)
print(f"ROC AUC Score for Decision Tree trained on Synthetic = {synthetic_roc_auc*100:.2f}%")
```

## Plots to show Performance

### Precision-Recall Curve

```python
vanilla_prerec_curve = precision_recall_curve(Y_test, pred_vanilla)
undersampled_prerec_curve = precision_recall_curve(Y_test, pred_undersampled)
synthetic_prerec_curve = precision_recall_curve(Y_test, pred_synthetic)

plt.figure(figsize=(7,4))
plt.plot(vanilla_prerec_curve[0], vanilla_prerec_curve[1], label='Vanilla Trained')
plt.plot(undersampled_prerec_curve[0], undersampled_prerec_curve[1], label='Undersampled Trained')
plt.plot(synthetic_prerec_curve[0], synthetic_prerec_curve[1], label='Synthetic Trained')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid()
plt.show()
```

### ROC AUC

```python
vanilla_roc_curve = roc_curve(Y_test, pred_vanilla)
undersampled_roc_curve = roc_curve(Y_test, pred_undersampled)
synthetic_roc_curve = roc_curve(Y_test, pred_synthetic)

plt.figure(figsize=(7,4))
plt.plot(vanilla_roc_curve[0], vanilla_roc_curve[1], label='Vanilla Trained')
plt.plot(undersampled_roc_curve[0], undersampled_roc_curve[1], label='Undersampled Trained')
plt.plot(synthetic_roc_curve[0], synthetic_roc_curve[1], label='Synthetic Trained')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid()
plt.show()
```
